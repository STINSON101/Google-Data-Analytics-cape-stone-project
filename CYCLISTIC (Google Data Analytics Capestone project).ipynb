```python
#__________________________PREPARATION_____________________________#
#importing pandas and reading 12 csv files 
import pandas as pd

df1 = pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202205.csv")
df2 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202206.csv")
df3 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202207.csv")
df4 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202208.csv")
df5 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202209.csv")
df6 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202210.csv")
df7 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202211.csv")
df8 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202212.csv")
df9 =  pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202301.csv")
df10 = pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202302.csv")
df11 = pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202303.csv")
df12 = pd.read_csv("C:/Users/stins/cyclistic_tripdata/XLX/202304.csv")
```

```python
#combining all 12 datasets into one dataset.
data = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12])

#_________________________Exploration__________________________#

data.info() # summary about the dataframe.
```

```python
# changing the datatype of ('started at' and 'ended at') date from object to datetime64.
data['started_at'] = pd.to_datetime(data['started_at'])
data['ended_at'] = pd.to_datetime(data['ended_at'])
```

```python
data.info()
```

```python
#from the column extract day, month and year.
data['year'] = data['started_at'].dt.year
data['month'] = data['started_at'].dt.month_name()
data['week_day'] = data['started_at'].dt.day_name()

data.head(10) # checking first 10 rows of the dataframe
```

```python
data.info()
```

```python
#______________________________Cleaning____________________________#
# removing unwanted columns from dataframe using the drop() function
data = data.drop(['start_station_id','end_station_id'], axis = 1)
```

```python
data = data.drop([' '], axis = 1) #removed an unnamed column.
```

```python
data = data.drop(['start_lat','start_lng','end_lat','end_lng'], axis = 1) #removed lats and longs.
```

```python
data = data.reset_index(drop = True) #reset the index back to the default 0, 1, 2 etc indexes.
```

```python
#droping all null values in dataframe.
data = data.dropna()
```

```python
data = data.reset_index(drop = True)
```

```python
data.to_csv("C:/Users/stins/cyclistic_tripdata/XLX/Cyclistic_cleaned_data.csv")
```
